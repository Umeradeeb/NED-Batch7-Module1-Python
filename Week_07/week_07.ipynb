```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean()) 
```

```python
df.tail()
```

```python
df['Type'] = df['Type'].fillna(df['Type'].mode()[0])
```

```python
df['Content Rating'] = df['Content Rating'].fillna(df['Content Rating'].mode()[0])
```

```python
df['Current Ver'] = df['Current Ver'].replace('Varies with device', np.nan)
```

```python
df['Android Ver'] = df['Android Ver'].replace('Varies with device', np.nan)
```

```python
df['Current Ver'] = pd.to_numeric(df['Current Ver'], errors='coerce').astype(float)
```

```python
df['Current Ver'] = df['Current Ver'].fillna(df['Current Ver'].mode()[0])
```

```python
df['Android Ver'] = df['Android Ver'].fillna(df['Android Ver'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Reviews'] = pd.to_numeric(df['Reviews'], errors='coerce').astype('Int64')
```

```python
df['Size'] = df['Size'].replace('Varies with device', np.nan)
```

```python
df['Size'] = df['Size'].replace('1,000+', np.nan)
```

```python
df['Size'] = df['Size'].str.replace('M', '').str.replace('k', '').astype(float)
```

```python
df['Size'] = df['Size'].fillna(df['Size'].mean())
```

```python
df['Size']
```

```python

df['Size'] = (df['Size'] - df['Size'].min()) / (df['Size'].max() - df['Size'].min())
```
